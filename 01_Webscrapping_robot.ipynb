```python
import requests

# List of websites to check robots.txt files
websites = [
    "https://www.google.com/robots.txt",
    "https://www.amazon.com/robots.txt",
    "https://www.wikipedia.org/robots.txt",
    "https://dailynous.com/robots.txt"
]

# Fetching and printing robots.txt content for each website
robots_data = {}
for website in websites:
    try:
        response = requests.get(website)
        if response.status_code == 200:
            robots_data[website] = response.text
        else:
            robots_data[website] = f"Failed to fetch. HTTP Status Code: {response.status_code}"
    except Exception as e:
        robots_data[website] = f"Error: {e}"


# Print the robots.txt contents
for site, content in robots_data.items():
    print(f"Robots.txt for {site}:\n{'-'*40}\n{content}\n{'-'*40}\n")
```
